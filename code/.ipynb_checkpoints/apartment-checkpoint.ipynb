```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

SEED = 42
AZIMUTH = -111
ELEVATION = 58
ROLL = 0
pd.options.display.max_columns = None
```

```python
!ls "data/apartment"
```

```python
data = pd.read_csv("data/apartment/apartments_for_rent_classified_100K.csv",
                   sep=";", encoding='cp1252', on_bad_lines="skip", low_memory=False)
print("Shape:", data.shape)
data.head()
```

```python
# Delete some columns
data = data.drop(["pets_allowed", "amenities", "address", "title", "body",
                  "currency", "price_display", "id", "cityname"], axis=1)

for col in data.columns:
    print(f"Column {col} has {len(data[col].unique())} modalities")
```

```python
data["category"] = data["category"].apply(lambda x:x.split('/')[-1] if x is not np.nan else "other")
data["bathrooms"] = data["bathrooms"].fillna(1)
data.isna().sum().to_frame().T
```

```python
data = data.dropna()
data.isna().sum().to_frame().T
```

```python
data.head()
```

```python
for col in data.select_dtypes("object").columns:
    data[col] = LabelEncoder().fit_transform(data[col])
data.head()
```

## Clustering

```python
cols = data.columns.difference(["category", "cityname"]).tolist()
set(cols)
```

```python
km =  KMeans(n_clusters=6, n_init=30, random_state=SEED)
km.fit(data[cols].copy())
```

```python
pca = PCA(n_components = 3);
X_reduced = pca.fit_transform(data[cols]);

x_axis = [val[0] for val in X_reduced];
y_axis = [val[1] for val in X_reduced];
z_axis = [val[2] for val in X_reduced];
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5), facecolor="w");

ax = fig.add_subplot(1, 2, 1, projection='3d');
plt.title('7-means');
ax.scatter(x_axis, y_axis, z_axis, c=km.labels_);
ax.view_init(ELEVATION, AZIMUTH, ROLL)

ax = fig.add_subplot(1, 2, 2, projection='3d');
plt.title('True');
ax.scatter(x_axis, y_axis, z_axis, c=data["target"]);
ax.view_init(ELEVATION, AZIMUTH, ROLL)

plt.show()
```
