```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

pd.options.display.max_columns = None

SEED = 42
AZIMUTH = -111
ELEVATION = 58
ROLL = 0
```

```python
!ls "data/"
```

```python
data = pd.read_csv("data/support2csv/support2.csv", encoding="utf-8")
print("Shape:", data.shape)
data.head()
```

```python
# Delete "scoma", "sps", "surv2m", "surv6m" because they are predicted by a model.
# and other because too many nan values
# "dnrday" same as dnr
data = data.drop(["scoma", "sps", "surv2m", "surv6m", "totmcst", "dnrday",
                  "alb", "glucose", "bun", "urine", "adlp", "adls", "sfdm2", "bili"], axis=1)
data["age"] = data["age"].astype("int8")
data["edu"] = np.ceil((data["edu"] / 4).replace({np.nan:-1}))
data["income"] = data["income"].replace({ np.nan:0, 'under $11k':1, '$11-$25k':2,
                                         '$25-$50k':3, '>$50k':4})
data["race"] = data["race"].replace({np.nan:0, 'other':0, 'white':1, 'black':2, 'hispanic':3, 'asian':4})
data["dnr"] = data["dnr"].replace({np.nan:0, 'no dnr':0, 'dnr after sadm':1, 'dnr before sadm':2})
data = data.fillna(0)
# data["charges"] = data["charges"].fillna(0)
# data["totcst"] = data["totcst"].fillna(0)
# data["avtisst"] = data["avtisst"].fillna(0)
# data[["prg2m", "prg6m"]] = data[["prg2m", "prg6m"]].filna(0)
# data["meanbp"] = data["meanbp"].fillna(0)
# data["wblc"] = data["wblc"].fillna(0)

data.head()
```

```python
# Encode categorial features
for col in data.select_dtypes(include="object").columns:
    data[col] = LabelEncoder().fit_transform(data[col])

data.head()
```

## Clustering

```python
cols = data.columns.difference(["dzgroup", "death", "hospdead"])
X = data[cols].values
```

```python
km =  KMeans(n_clusters=2, n_init=50, random_state=SEED)
km.fit(X.copy())
```

```python
pca = PCA(n_components = 3);
X_reduced = pca.fit_transform(X.copy());

x_axis = [val[0] for val in X_reduced];
y_axis = [val[1] for val in X_reduced];
z_axis = [val[2] for val in X_reduced];X_reduced[:, 0]
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5), facecolor="w");

ax = fig.add_subplot(1, 1, 1, projection='3d');
plt.title('4-means');
ax.scatter(x_axis, y_axis, z_axis, c=km.labels_);
ax.view_init(ELEVATION, AZIMUTH, ROLL)

# ax = fig.add_subplot(1, 2, 2, projection='3d');
# plt.title('True');
# ax.scatter(x_axis, y_axis, z_axis, c=data["page 1 (main category)"]);
# ax.view_init(ELEVATION, AZIMUTH, ROLL)

plt.show()
```
