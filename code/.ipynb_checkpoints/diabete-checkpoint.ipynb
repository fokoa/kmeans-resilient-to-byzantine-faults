```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

SEED = 42
AZIMUTH = -111
ELEVATION = 58
ROLL = 0
pd.options.display.max_columns = None
```

```python
!ls "data/diabete/"
```

```python
data = pd.read_csv("data/diabete/diabetic_data.csv")
print("Shape:", data.shape)
data.head()
```

```python
# Delete some columns
data = data.drop(["max_glu_serum", "A1Cresult", "encounter_id", "patient_nbr", "weight",
                  "payer_code", ], axis=1)

for col in data.columns:
    print(f"Column {col} has {len(data[col].unique())} modalities")
```

```python
for col in data.select_dtypes("object").columns:
    data[col] = data[col].replace({"?": "Other", "Unknown/Invalid":"Other"}) 
```

```python
data["age"] = data["age"].replace({'[0-10)':0, '[10-20)':1, '[20-30)':2, '[30-40)':3,
                                    '[40-50)':4, '[50-60)':5, '[60-70)':6, '[70-80)':7,
                                    '[80-90)':8, '[90-100)':9})
data["medical_specialty"] = data["medical_specialty"].apply(lambda x: x.split("-")[0] if x is not np.nan else x)
data["medical_specialty"] = data["medical_specialty"].replace({"Hematology/Oncology":"Oncology",
                                         "Obsterics&Gynecology":"Obstetrics"   })
tmp = data["medical_specialty"].value_counts()
data.loc[data["medical_specialty"].isin(tmp[tmp.values<40].index.values), "medical_specialty"] = "Other"
```

```python
values = set(data["diag_1"]).union(set(data["diag_2"])).union(set(data["diag_3"]))
replaces = {}
for i, val in enumerate(values):
    replaces[val] = i
data["diag_1"] = data["diag_1"].replace(replaces)
data["diag_2"] = data["diag_2"].replace(replaces)
data["diag_3"] = data["diag_3"].replace(replaces)
```

```python
for col in data.select_dtypes("object").columns:
    data[col] = LabelEncoder().fit_transform(data[col])
```

```python
print(data.shape)
data.head()
```

## Clustering

```python
target = "readmitted"
cols = data.columns.difference(["readmitted"]).tolist()
len(cols)
```

```python
pca = PCA(n_components = 3);
X_reduced = pca.fit_transform(data[cols]);

x_axis = [val[0] for val in X_reduced];
y_axis = [val[1] for val in X_reduced];
z_axis = [val[2] for val in X_reduced];
```

```python
km =  KMeans(n_clusters=3, n_init=30, random_state=7)
km.fit(X_reduced)#(data[cols].copy())
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5), facecolor="w");

ax = fig.add_subplot(1, 2, 1, projection='3d');
plt.title('7-means');
ax.scatter(x_axis, y_axis, z_axis, c=km.labels_);
ax.view_init(ELEVATION, AZIMUTH, ROLL)

ax = fig.add_subplot(1, 2, 2, projection='3d');
plt.title('True');
ax.scatter(x_axis, y_axis, z_axis, c=data[target]);
ax.view_init(ELEVATION, AZIMUTH, ROLL)

plt.show()
```
