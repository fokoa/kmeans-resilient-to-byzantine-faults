```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder

SEED = 42
AZIMUTH = -111
ELEVATION = 58
ROLL = 0
```

```python
!ls "data/clickstream"
```

```python
data = pd.read_csv("data/clickstream/e-shop clothing 2008.csv", sep=";")
print("Shape:", data.shape)
data.head()
```

```python
for col in data.columns:
    print(f"Column {col} has {len(data[col].unique())} modalities")
```

```python
# Delete some columns
data = data.drop(["year", "page 2 (clothing model)"], axis=1)
data.head()
```

## Clustering

```python
target = "page 1 (main category)"
cols = data.columns.difference([target, "country"]).tolist()
```

```python
km =  KMeans(n_clusters=4, n_init=30, random_state=SEED)
km.fit(data[cols].copy())
```

```python
pca = PCA(n_components = 3);
X_reduced = pca.fit_transform(data[cols]);

x_axis = [val[0] for val in X_reduced];
y_axis = [val[1] for val in X_reduced];
z_axis = [val[2] for val in X_reduced];
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5), facecolor="w");

ax = fig.add_subplot(1, 2, 1, projection='3d');
plt.title('4-means');
ax.scatter(x_axis, y_axis, z_axis, c=km.labels_);
ax.view_init(ELEVATION, AZIMUTH, ROLL)

ax = fig.add_subplot(1, 2, 2, projection='3d');
plt.title('True');
ax.scatter(x_axis, y_axis, z_axis, c=data["page 1 (main category)"]);
ax.view_init(ELEVATION, AZIMUTH, ROLL)

plt.show()
```
